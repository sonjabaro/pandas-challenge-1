## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the 
# subtotal for each line using the unit_price
# and the qty
#first create the subtotal column with the calculation in it
df["line_subtotal"]= df["unit_price"] * df["qty"]
#then put the new column with the others and checked output with .head
df[["unit_price", "qty", "line_subtotal"]].head(2)
```

```python
# Create a column for shipping price.
#create colunm for total weight
#had to find out weight to determine shipping price
total_weight = df ["unit_weight"] * df["qty"]
df["total_weight"] = total_weight

# Assume a shipping price of $7 per pound
# for orders over 50 pounds and $10 per
# pound for items 50 pounds or under.
#Made a function to calculate shipping weight
def calculate_shipping_price(weight):
    if weight > 50:
        return weight * 7
    else:
        return weight * 10
    #after defining the function I created the column for the shipping price using the total weight for defined above and using apply method to apply this and the shipping price function to each row
df["shipping_price"]=df["total_weight"].apply(calculate_shipping_price)
#Displayed new table with the updated added column
df[["unit_price", "unit_weight", "qty", "total_weight",  "shipping_price"]].head(3)
```

```python
# Create a column for the total price
# using the subtotal and the shipping price
# along with a sales tax of 9.25%
#Define sales tax against the subtotal
sales_tax = 1.0925
#calculate the total price (line subtotal + shipping price ) * tax  
df["total_price"]=(df["line_subtotal"] + df["shipping_price"]) * sales_tax
#make the number to use 2 digit cents using round method
df["total_price"] = df["total_price"].round(2)
#show all columns
df[["line_subtotal","shipping_price","total_price"]].head(3)
```

```python
# Create a column for the cost
# of each line using unit cost, qty, and
# shipping price (assume the shipping cost
# is exactly what is charged to the client).

#create new column and calculation (cost*qty)and add the shipping cost (note our study group used this formula, however would we not include tax?)
df["line_cost"]= df["unit_cost"]*df["qty"] + df["shipping_price"]
df[["line_cost","unit_cost","qty","shipping_price"]].head(3)
```

```python
# Create a column for the profit of
# each line using line cost and line price
# define profit margin
df["line_profit"] = df["total_price"] - df["line_cost"]

#list columns
df[["line_profit", "total_price", "line_cost"]].head(3)

```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity
# spend? Check your work from Part 1 for client ids.
#create function for top spend

#create a function to locate the id and the total price -must add sum method to get total

def find_total_price(client_id):
    return df.loc[df["client_id"]== client_id, "total_price"].sum()

#ids from previous work
top_5_client_ids = [33615,66037,46820,38378,24741]

#pass in info into a loop and format
for client_id in top_5_client_ids:
    total_spent = find_total_price(client_id)
    print(f"{client_id}: ${total_spent:.2f}")

```

```python
# Create a summary DataFrame showing the totals for the
# for the top 5 clients with the following information:
# total units purchased, total shipping price,
# total revenue, and total profit. Sort by total profit.

#create functions
# total purchased by client
def total_units_purchased(client_id):
    return df.loc[df["client_id"] == client_id, "qty"].sum()
# using applymethod for each line of purchase determine total weight (total weight = unit *qty)
#using apply method apply the function for the shipping price
# return the shipping price by line and sum all lines
def total_shipping_price(client_id):
    total_weight = df.loc[df["client_id"] == client_id,"unit_weight"]* df.loc[df['client_id'] == client_id, 'qty']
    shipping_price =total_weight.apply(lambda weight:weight *7 if weight >50  else weight * 10)
    return shipping_price.sum()
    
# total revenue =(total price summed by each line)
def total_revenue(client_id):
    return df.loc[df["client_id"] == client_id,"total_price"].sum()
#total profit by client - identify line cost, then subtract cost from total price
def total_profit(client_id):
    line_cost = df.loc[df["client_id"] == client_id, "unit_cost"] * df.loc[df['client_id'] == client_id, 'qty'] + df.loc[df["client_id"]== client_id, "shipping_price"]
    line_profit = df.loc[df["client_id"] == client_id, "total_price"] - line_cost
    return line_profit.sum()

    

#identify top 5 ids
top_5_client_ids = [33615,66037,46820,38378,24741]
#define client data variable
client_data = []
#run loop
for client_id in top_5_client_ids:
    total_units_purchased_val = total_units_purchased(client_id)
    total_shipping_price_val = total_shipping_price(client_id)
    total_revenue_val = total_revenue(client_id)
    total_profit_val = total_profit(client_id)
    #rename columns
    client_info = {
        "client_id":client_id,
        "qty": total_units_purchased_val,
        "shipping_price": total_units_purchased_val,
        "total_price": total_revenue_val,
        "line_profit": total_profit_val
    }
    client_data.append(client_info)
    
summary_client_data_df = pd.DataFrame(client_data)
#sort by value of profit in decending order (highest to lowest)
summary_client_data_df = summary_client_data_df.sort_values(by = "line_profit", ascending = False)
summary_client_data_df
```

```python
# Format the data and rename the columns
# to names suitable for presentation.
# Currency should be in millions of dollars.

#convert currency first using/1000000


summary_client_data_df["total_price"] = summary_client_data_df["total_price"] / 1_000_000  # Convert to millions
summary_client_data_df["shipping_price"] = summary_client_data_df["shipping_price"] / 1_000_000  # Convert to millions
summary_client_data_df["line_profit"] = summary_client_data_df["line_profit"] / 1_000_000 
#Rename columns using rename method
summary_client_data_df = summary_client_data_df.rename(columns={
    "total_price": "Total Price (Millions)",
    'shipping_price': 'Shipping Price (Millions)',
    'qty': 'Quantity',
    'client_id': 'Client ID',
    "line_profit": "Total Profit(Millions)"
})
summary_client_data_df



```

```python
# Sort the updated data by "Total Profit" form highest to lowest
#use sort values method, ascending is false since it is to be highest to lowest
summary_client_data_df=summary_client_data_df.sort_values(by = "Total Profit(Millions)", ascending=False)
summary_client_data_df
```

```python
#Additional analysis I wanted to do - find the percentage that each top 5 client represented for the business.

# Sort the DataFrame by Total Price (Millions) in descending order and select the top 5 clients
top_5_clients = summary_client_data_df.sort_values(by='Total Price (Millions)', ascending=False)

# Calculate the total sales of the top 5 clients
total_sales_top_5_clients = top_5_clients['Total Price (Millions)'].sum()

# Calculate the percentage breakout for each client
top_5_clients['Percentage'] = (top_5_clients['Total Price (Millions)'] / total_sales_top_5_clients) * 100

# Display the results
print(f"Total Sales of Top 5 Clients: {total_sales_top_5_clients:.2f} Million Dollars")
print("\nTop 5 Clients Breakout by Percentage:")
print(top_5_clients[['Client ID', 'Total Price (Millions)', 'Percentage']].to_string(index=False))
```
